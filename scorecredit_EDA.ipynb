## Matriz de correlacin entre las variables numricas

```python
# Correlacin entre variables numricas
plt.figure()
sns.heatmap(df[num_vars].corr(), annot=True, cmap="coolwarm", center=0)
plt.title("Matriz de correlacin")
plt.show()
```

Al analizar la matriz de correlacin se observa que algunas variables presentan relaciones fuertes entre s. Por ejemplo, Interest_Rate, Num_of_Loan, Outstanding_Debt, Num_Credit_Inquiries y Num_of_Delayed_Payment muestran correlaciones positivas relativamente altas (entre 0.5 y 0.64), lo que indica que aportan informacin similar y que incluirlas todas en el modelo podra generar redundancia.

En contraste, Credit_History_Age presenta correlaciones negativas con estas variables (entre 0.42 y 0.63), lo cual lo convierte en un predictor complementario que aade una perspectiva distinta sobre el riesgo crediticio.

Por otro lado, Credit_Utilization_Ratio muestra correlaciones muy bajas con el resto de las variables, lo que lo convierte en un indicador prcticamente independiente y especialmente til para enriquecer el modelo.

Con base en este anlisis, se seleccionaron las siguientes variables para el scorecard: **Num_Credit_Card, Num_of_Loan, Delay_from_due_date, Num_of_Delayed_Payment, Num_Credit_Inquiries, Credit_Mix, Outstanding_Debt, Credit_History_Age y Payment_of_Min_Amount, Interest_Rate, Credit_Utilization_Ratio**. Estas variables ofrecen un balance adecuado entre historial de pago, nivel de endeudamiento, comportamiento crediticio reciente y decisiones de pago, evitando duplicidad excesiva de informacin.
