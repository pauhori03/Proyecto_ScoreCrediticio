## Matriz de correlación entre las variables numéricas

```python
# Correlación entre variables numéricas
plt.figure()
sns.heatmap(df[num_vars].corr(), annot=True, cmap="coolwarm", center=0)
plt.title("Matriz de correlación")
plt.show()
```

Al analizar la matriz de correlación se observa que algunas variables presentan relaciones fuertes entre sí. Por ejemplo, Interest_Rate, Num_of_Loan, Outstanding_Debt, Num_Credit_Inquiries y Num_of_Delayed_Payment muestran correlaciones positivas relativamente altas (entre 0.5 y 0.64), lo que indica que aportan información similar y que incluirlas todas en el modelo podría generar redundancia.

En contraste, Credit_History_Age presenta correlaciones negativas con estas variables (entre –0.42 y –0.63), lo cual lo convierte en un predictor complementario que añade una perspectiva distinta sobre el riesgo crediticio.

Por otro lado, Credit_Utilization_Ratio muestra correlaciones muy bajas con el resto de las variables, lo que lo convierte en un indicador prácticamente independiente y especialmente útil para enriquecer el modelo.

Con base en este análisis, se seleccionaron las siguientes variables para el scorecard: **Num_Credit_Card, Num_of_Loan, Delay_from_due_date, Num_of_Delayed_Payment, Num_Credit_Inquiries, Credit_Mix, Outstanding_Debt, Credit_History_Age y Payment_of_Min_Amount, Interest_Rate, Credit_Utilization_Ratio**. Estas variables ofrecen un balance adecuado entre historial de pago, nivel de endeudamiento, comportamiento crediticio reciente y decisiones de pago, evitando duplicidad excesiva de información.

## Histogramas Segmentados

Se procedió a realizar un análisis individual de cada una de las variables seleccionadas con el fin de evaluar su poder discriminante respecto al Credit_Score real presente en el dataset. Para ello, se construyeron histogramas segmentados por las categorías Good, Standard y Poor, lo que permite observar cómo se distribuye cada grupo dentro de los diferentes rangos de valores de cada variable.

Este análisis es fundamental porque muestra visualmente si existe una separación clara entre las clases. En otras palabras, permite identificar si, por ejemplo, los clientes con Good score tienden a concentrarse en ciertos rangos de una variable (como baja utilización de crédito o pocos retrasos de pago), mientras que los clientes con Poor score se concentran en rangos opuestos.

De esta manera, el análisis gráfico respalda la elección de las variables para el modelo de scorecard, ya que confirma qué tan útiles son para diferenciar el riesgo crediticio y qué rangos pueden convertirse posteriormente en bins con puntajes asignados en el modelo de Excel.

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df, x="Num_Credit_Card", hue="Credit_Score", bins=30, multiple="stack",palette=["#f4547c", "#fcd856", "#51bfe2"])
plt.title("Number of Credit Cards by Credit Score")
plt.show()
```

El histograma muestra que la mayoría de las personas concentran entre 4 y 7 tarjetas de crédito, siendo este el rango más frecuente. Dentro de este rango predominan los clientes con score Standard, seguidos por los de Poor, mientras que los de Good aparecen en menor proporción. Los clientes con score Good se concentran más en niveles bajos y moderados (1 a 5 tarjetas), mientras que conforme aumenta el número de tarjetas (8 o más) son los clientes con score Standard y Poor quienes predominan, lo que sugiere que la acumulación excesiva de tarjetas está más asociada con un mayor riesgo crediticio.

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df, x="Num_of_Delayed_Payment", hue="Credit_Score", bins=30,multiple="stack", palette=["#f4547c", "#fcd856", "#51bfe2"])
plt.title("Number of Delayed Payment by Credit Score")
plt.show()
```

Se muestra que los clientes con pocos o nulos retrasos (0–5 pagos) están más asociados a la categoría Good y en menor medida a Standard. Esto indica que el número de retrasos de pago es un predictor fuerte del riesgo: a mayor cantidad de atrasos, mayor probabilidad de pertenecer a la categoría Poor.

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df, x="Outstanding_Debt", hue="Credit_Score", bins=30, multiple="stack",palette=["#f4547c", "#fcd856", "#51bfe2"])
plt.title("Outstanding Debt by Credit Score")
plt.show()
```

## Conclusión

El análisis exploratorio permitió identificar las variables más relevantes para la construcción del scorecard de crédito. A través de la matriz de correlación se detectó que algunas variables presentaban una alta relación entre sí, lo que implicaba redundancia de información. Esto justificó la selección de nueve predictores clave: **Num_Credit_Card, Num_of_Loan, Delay_from_due_date, Num_of_Delayed_Payment, Num_Credit_Inquiries, Credit_Mix, Outstanding_Debt, Credit_History_Age y Payment_of_Min_Amount, Interest_Rate, Credit_Utilization_Ratio**, los cuales ofrecen un balance adecuado entre historial de pago, nivel de endeudamiento, hábitos de pago y diversidad de productos financieros.

El análisis gráfico de cada variable frente al Credit_Score real evidenció patrones consistentes: los clientes con mejor calificación se concentran en valores asociados a menor riesgo (menos retrasos, menor deuda, menor número de consultas, mejor combinación de créditos, tasas más bajas y mayor antigüedad crediticia), mientras que los clientes con calificación baja tienden a agruparse en los extremos opuestos, mostrando comportamientos de mayor vulnerabilidad financiera.

En conjunto, estos resultados confirman que las variables seleccionadas poseen capacidad discriminante y constituyen una base sólida para el diseño de los bins y la asignación de puntajes en el modelo de Excel, asegurando que el scorecard capture de manera efectiva la diferencia entre clientes de alto y bajo riesgo.
